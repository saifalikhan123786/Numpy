## Advanced Operations with Bussiness Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vec1=np.array([1,2,3,4,5])
vec2=np.array([5,6,7,8,9])

angle= np.arccos(np.dot(vec1,vec2) /(np.linalg.norm(vec1) * np.linalg.norm(vec2)))
print(angle)
```
